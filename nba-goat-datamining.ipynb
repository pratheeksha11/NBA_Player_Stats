```python
df = pd.read_csv('/kaggle/input/nba-players-data/all_seasons.csv')
df
```

```python
import pandasql as psql
```

```python
query = "SELECT * FROM df WHERE player_name = 'LeBron James'"

lebron_james = psql.sqldf(query, locals())

lebron_james
```

```python
lebron_james.to_csv('/kaggle/working/lebron_james.csv', index=False)
print('file saved')
```
